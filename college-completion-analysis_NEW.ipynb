```python
# import training and test dataset
train_df = pd.read_csv('/kaggle/input/us-college-completion-rate-analysis/train.csv')
test_df = pd.read_csv('/kaggle/input/us-college-completion-rate-analysis/x_test.csv')
```

```python
train_df # display training dataset for observation
```

```python
# simply check if any missing value for training dataset
col_list = ['Completion_rate', 'Tuition_in_state', 'Tuition_out_state', 'Faculty_salary', 'Pell_grant_rate', \
            'SAT_average', 'ACT_50thPercentile', 'pct_White', 'pct_Black', 'pct_Hispanic', 'pct_Asian', \
           'Parents_middlesch', 'Parents_highsch', 'Parents_college']

print('Training Dataset Empty Values Check:')
for i in col_list:
    empty_value = train_df[i].isnull().sum()
    print(f"Missing value in column {i}:",empty_value)
```

```python
test_df # display testing dataset for observation
```

```python
# simply check if any missing value for testing dataset
col_list = ['Tuition_in_state', 'Tuition_out_state', 'Faculty_salary', 'Pell_grant_rate', \
            'SAT_average', 'ACT_50thPercentile', 'pct_White', 'pct_Black', 'pct_Hispanic', 'pct_Asian', \
           'Parents_middlesch', 'Parents_highsch', 'Parents_college']

print('Testing Dataset Empty Values Check:')
for i in col_list:
    empty_value = test_df[i].isnull().sum()
    print(f"Missing value in column {i}:",empty_value)
```

```python
# select relevant features from training dataset and display the x and y training data
train_data = train_df[['Tuition_out_state', 'Parents_college', 'pct_White', 'SAT_average']]
x_train = train_data.values
y_train = train_df['Completion_rate']
x_train, y_train
```

```python
# fit regression model and display scores
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
print(model.score(x_train, y_train))
```

```python
# use the model to make predictions for testing dataset
test_data = test_df[['Tuition_out_state', 'Parents_college', 'pct_White', 'SAT_average']]
x_test = test_data.values
predictions = model.predict(x_test)
print(predictions)
```

```python
# observe model coefficients for any possible improvements in milestone 2
pd.DataFrame(model.coef_, test_data.columns, columns = ['Coeff'])
```

```python
# create submission dataset and display to final check
submission = pd.DataFrame.from_dict({'Completion_rate': predictions})
submission
```

```python
# save submission file
submission.to_csv('submission.csv', index = True, index_label = "id") 
```
